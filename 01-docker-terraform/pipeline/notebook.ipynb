```python
parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    prefix + '/yellow_tripdata_2021-01.csv.gz',
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
engine = create_engine('postgresql://root:root@localhost:5434/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, con=engine, name='yellow_taxi_data'))
```

```python
df.head(0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
    url,
    chunksize = 100000,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True
)
```

```python
df = next(df_iter)
```

```python
df
```

```python
from tqdm.auto import tqdm
```

```python
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )
```

```python
print(type(df_iter))
for df_chunk in df_iter:
    print("Chunk received:", df_chunk.shape)
    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```
