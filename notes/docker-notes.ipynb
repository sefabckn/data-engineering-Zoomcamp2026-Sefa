### **Key Concepts**

#### ğŸ—ï¸ **Image vs Container**
- **Image**: Blueprint (like a recipe)
- **Container**: Running instance (like a cooked meal)

```bash
# Build image
docker build -t my-app:1.0 .

# Run container from image
docker run -d -p 5432:5432 my-app:1.0
```

#### ğŸ“š **Layer Caching (The Speedup Trick)**
```dockerfile
FROM python:3.13-slim          # Layer 1: Base image
COPY requirements.txt .         # Layer 2: Dependencies
RUN pip install -r requirements # Layer 3: Install (SLOW)
COPY app.py .                   # Layer 4: Code
```

**Why order matters:**
- Dependencies change rarely â†’ Cache stays valid
- Code changes often â†’ But dependencies are already cached
- **Result**: 30s rebuild instead of 5 minutes! âš¡

#### ğŸ”— **Docker Compose: Multiple Containers**
```yaml
services:
  postgres:    # Database service
  pgadmin:     # UI for database
  app:         # Your application
```

**Magic:** Services communicate by name (`postgres:5432`), volumes persist data, networks are automatic.

### **Core Building Blocks**

## ğŸ’¡ Key Takeaways

### **Docker**
1. **Think in containers**: One container = one concern (DB, app, cache)
2. **Layer efficiently**: Put changing stuff last
3. **Use compose locally**: Test full stack before deploying
4. **Never hardcode secrets**: Use environment variables

### **Terraform**
1. **Code your infrastructure**: No more clicking in dashboards
2. **Use variables**: Make code reusable across environments
3. **Plan before apply**: Read the plan output carefully!
4. **State files are gold**: Treat `terraform.tfstate` like a database backup
5. **Version your infra**: Commit `.tf` files to Git

### **Data Engineering Perspective**
- **Docker** = reproducible applications
- **Terraform** = reproducible infrastructure
- **Together** = fully reproducible data pipelines
- **Result** = Reliable, scalable, maintainable systems âœ¨
